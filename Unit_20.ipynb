### Unit. 20 심사 문제

```python
start, end = map(int, input().split())

for i in range(start, end + 1):
    if i % 5 == 0 and i % 7 == 0:
        print("FizzBuzz")
    elif i % 7 == 0:
        print("Buzz")
    elif i % 5 == 0:
        print("Fizz")
    else:
        print(i)
```
